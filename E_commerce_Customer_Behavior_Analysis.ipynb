```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

pd.set_option('display.float_format', lambda x: '{:.3f}'.format(x))
np.set_printoptions(suppress=True)
```

```python
df=pd.read_csv('/content/E-commerce Customer Behavior - Sheet1.csv')
df.head()
```

```python
df.columns=df.columns.str.replace(' ','_').str.lower()
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df['satisfaction_level'].unique()
```

```python
df['satisfaction_level'].mode()
```

```python
df['satisfaction_level']=df['satisfaction_level'].fillna('Satisfied')
```

```python
df.duplicated().sum()
```

```python
df=df.drop('customer_id',axis=1)
```

```python
df.info()
```

```python
int_cols=df.select_dtypes(include='int64').columns
```

```python
cols_to_encode=df.select_dtypes(include='object').columns
for c in cols_to_encode:
  print(c,'-->',df[c].unique())
```

```python
df=pd.get_dummies(df,columns=['membership_type'],drop_first=True)
```

```python
df['gender']=df['gender'].map({'Female':0,'Male':1})

df['city']=df['city'].map({'New York':0,'Los Angeles':1,'Chicago':2,'San Francisco':3,'Miami':4,'Houston':5})

df['satisfaction_level']=df['satisfaction_level'].map({'Satisfied':0,'Neutral':1,'Unsatisfied':2})
```

```python
df.info()
```

```python
bool_cols=df.select_dtypes(include='bool').columns
df[bool_cols]=df[bool_cols].astype(int)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df[int_cols]=scaler.fit_transform(df[int_cols])
df.describe()
```

```python
X=df.drop('average_rating',axis=1)
y=df['average_rating']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=45)
```

```python
from sklearn.linear_model import LinearRegression
model_lr=LinearRegression()
model_lr.fit(X_train,y_train)
model_lr.score(X_test,y_test)
```

```python
from xgboost import XGBRegressor
model_xgb=XGBRegressor()
model_xgb.fit(X_train,y_train)
model_xgb.score(X_test,y_test)
```
